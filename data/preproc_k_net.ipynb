```python
import os
import re
import pandas as pd

base = "/Volumes/Local Disk/Code_Git/S3_code/seismic/k_net"

rows = []

pattern = re.compile(r"([A-Z]{3,4}\d{2,3})(\d{10})(?:\.(EW|NS|UD)(\d)?)?")

for root, dirs, files in os.walk(base):
    for fname in files:
        fpath = os.path.join(root, fname)
        ext = os.path.splitext(fname)[1]

        station = None
        timestamp = None
        comp = None
        channel = None

        m = pattern.search(fname)
        if m:
            station = m.group(1)
            timestamp = m.group(2)
            comp = m.group(3)
            channel = m.group(4)

        rows.append({
            "folder": os.path.basename(root),
            "filename": fname,
            "extension": ext,
            "station": station,
            "timestamp_raw": timestamp,
            "component": comp,
            "channel": channel,
            "file_type": (
                "waveform" if comp in ["EW", "NS", "UD"] else
                "plot_ps" if ext == ".gz" else
                "unknown"
            )
        })

df = pd.DataFrame(rows)
df.to_csv("knet_inventory_full.csv", index=False)

df.head(20)
```

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

# URL katalog JMA tahun 2015
url = "https://www.data.jma.go.jp/svd/eqev/data/daily_map/2015/index.html"

# Ambil halaman HTML
response = requests.get(url)
response.encoding = "utf-8"

soup = BeautifulSoup(response.text, "html.parser")

# Cari semua link harian (tiap hari punya daftar gempa)
links = soup.select("a")

events = []

for link in links:
    href = link.get("href")
    if href and href.endswith(".html"):
        day_url = "https://www.data.jma.go.jp/svd/eqev/data/daily_map/2015/" + href
        day_page = requests.get(day_url)
        day_page.encoding = "utf-8"
        day_soup = BeautifulSoup(day_page.text, "html.parser")

        # Cari tabel event
        tables = day_soup.find_all("table")

        for table in tables:
            rows = table.find_all("tr")
            for row in rows[1:]:  # skip header
                cols = [c.get_text(strip=True) for c in row.find_all("td")]
                if len(cols) >= 5:
                    events.append({
                        "origin_time": cols[0],
                        "latitude": cols[1],
                        "longitude": cols[2],
                        "depth_km": cols[3],
                        "magnitude_Mj": cols[4],
                        "region": cols[5] if len(cols) > 5 else None
                    })

# Simpan ke DataFrame
df = pd.DataFrame(events)

# Bersihkan format
df["origin_time"] = df["origin_time"].str.replace(" ", "")
df["depth_km"] = df["depth_km"].str.replace("km", "").astype(float)
df["magnitude_Mj"] = df["magnitude_Mj"].str.replace("M", "").astype(float)

# Simpan ke CSV
df.to_csv("JMA_2015_events.csv", index=False)

df.head(20)
```
