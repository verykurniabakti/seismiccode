```python
# merge csv files in a folder into one csv file named merge.csv 
import pandas as pd
import glob

def merge_csv(base_folder):
    csv_files = sorted(glob.glob(f"{base_folder}/chunk*.csv"))
    print("CSV ditemukan:", csv_files)

    if len(csv_files) == 0:
        raise ValueError("Tidak ada file CSV ditemukan. Periksa kembali path foldernya.")

    dfs = [pd.read_csv(f) for f in csv_files]
    merged = pd.concat(dfs, ignore_index=True)

    merged.to_csv("merge.csv", index=False)
    print("merge.csv berhasil dibuat dengan", len(merged), "baris")

merge_csv("/Volumes/Local Disk/Code_Git/S3_code/seismic/data")
```

```python
#merge hdf5 files in a folder into one hdf5 file named merge.hdf5
import h5py
import glob
import os
from tqdm import tqdm

def merge_hdf5(source_folder, output_folder):
    hdf5_files = sorted(glob.glob(f"{source_folder}/chunk*.hdf5"))
    print("HDF5 ditemukan:", len(hdf5_files))

    if len(hdf5_files) == 0:
        raise ValueError("Tidak ada file HDF5 ditemukan. Periksa path sumber.")

    output_path = os.path.join(output_folder, "merge.hdf5")

    with h5py.File(output_path, "w") as fout:
        grp = fout.create_group("data")

        # Hitung total trace untuk progress bar
        total_traces = 0
        for file in hdf5_files:
            with h5py.File(file, "r") as fin:
                total_traces += len(fin["data"].keys())

        print("Total trace yang akan digabung:", total_traces)

        pbar = tqdm(total=total_traces, desc="Merging HDF5", unit="trace")

        # Proses merge
        for file in hdf5_files:
            with h5py.File(file, "r") as fin:
                for key in fin["data"].keys():
                    fin["data"].copy(key, grp)
                    pbar.update(1)

        pbar.close()

    print(f"\nmerge.hdf5 selesai dibuat di: {output_path}")

merge_hdf5(
    source_folder="/Volumes/Local Disk/Code_Git/S3_code/seismic/data",
    output_folder="/Volumes/Extreme SSD/Stead"
)
```
